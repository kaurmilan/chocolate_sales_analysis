```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("chocolatesales.csv")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df['date'] =pd.to_datetime(df['date'], dayfirst = True)

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.month_name()

df.head(10)

```

```python
df['amount'] = (df['amount'].str.replace(',', '', regex=True).str.replace(r'\$','', regex=True).str.strip())
```

```python
df['amount'] = pd.to_numeric(df['amount'], errors='coerce')
```

```python
df.head()
df.info()
df.describe()
```

```python
df[['sales_person','country','product']].nunique()
```

```python
df['amount'].describe()
```

```python
df['country'].value_counts().sort_values(ascending=False).head(3)
```

```python
df['boxes_shipped'].dtype
```

```python
df.groupby('country')['amount'].sum().sort_values(ascending=False) #sales by country
```

```python
df.groupby('sales_person')['boxes_shipped'].sum().sort_values(ascending=False) #sales person perfomance
```

```python
df[['boxes_shipped', 'amount']].corr() #linear relation between boxes shipped and amount
```

```python
df.groupby(['year','month'])['amount'].sum().plot(kind='bar')
plt.show() #sales over times
```

```python
#best salesperson from each country
sales_summary=df.groupby(['country','sales_person'])['amount'].sum().reset_index()
best_sales = sales_summary.loc[sales_summary.groupby('country')['amount'].idxmax()].reset_index(drop=True)
print(best_sales)
```

```python
df.groupby('country')['amount'].sum()
```

```python
df.groupby('country')['amount'].sum().plot(kind='line')
plt.show() #sales by country
```

```python
#country-wise shipment comparison
country_sales = df.groupby('country')['boxes_shipped'].sum().sort_values()

plt.barh(country_sales.index, country_sales.values)
plt.xlabel('boxes_shipped')
plt.title('total boxes shipped by country')
plt.show()
```
